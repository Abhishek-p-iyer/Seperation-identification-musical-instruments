<H2> ~ Analysis of major instruments of the dataset ~</H2>

<H4>Librosa Library has been used to analyse the audio inputs. Matplotlib and seaborne libraries will be used to plot waveforms for the same.</H4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display as ipd

import librosa
import librosa.display
```

<H3>(I) Bass_clarinet sound analysis</H3>

```python
audio1 = 'instruments_samples\Bass_clarinet.mp3'
x , sr = librosa.load(audio1)
print(type(x), type(sr))
print(x.shape, sr)
```

<b>This code returns audio time series as a numpy array with a default sampling rate(sr) of 22KHZ mono. We can easily change this feature by resampling it at 44.1KHz (as given below).</b>

```python
librosa.load(audio1, sr=44100)
```

```python
ipd.Audio(audio1)
```

<H4>Plotting the audio array created above</H4>

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

<H4>Spectrogram</H4>
<H5>A spectrogram is a visual way of representing the signal strength, or “loudness”, of a signal over time at various frequencies present in a particular waveform.
A spectrogram is usually depicted as a heat map, i.e., as an image with the intensity shown by varying the color or brightness.</H5>

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz', cmap='jet')
plt.colorbar()
```
